# CSV Merge Utilities

Consolidate multiple CSV files from a folder into a single merged output.

## Setup

```python
import pandas as pd
import glob
import os

# Google Colab setup (uncomment if needed)
# from google.colab import drive
# drive.mount('/content/drive')

print("Setup complete.")
```

## Merge Function

```python
def merge_csv_files(folder_path, output_file, skip_pattern=None):
    """
    Merge all CSV files in a folder.
    
    Args:
        folder_path: Path to folder containing CSV files
        output_file: Output CSV path
        skip_pattern: Filename pattern to skip (e.g., 'merged')
    """
    csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
    
    if not csv_files:
        print(f"No CSV files found in: {folder_path}")
        return None
    
    print(f"Found {len(csv_files)} CSV file(s):")
    
    dfs = []
    for file_path in csv_files:
        file_name = os.path.basename(file_path)
        
        if skip_pattern and skip_pattern in file_name:
            print(f"  (skip) {file_name}")
            continue
        
        try:
            df = pd.read_csv(file_path)
            dfs.append(df)
            print(f"  + {file_name}: {len(df):,} rows")
        except Exception as e:
            print(f"  ! {file_name}: error - {e}")
    
    if not dfs:
        print("No valid CSV files to merge.")
        return None
    
    merged_df = pd.concat(dfs, ignore_index=True)
    
    merged_df.to_csv(output_file, index=False)
    
    print(f"\nMerge complete:")
    print(f"  Total rows: {len(merged_df):,}")
    print(f"  Columns: {len(merged_df.columns)}")
    print(f"  Output: {output_file}")
    
    return merged_df
```

## Batch Merge Multiple Folders

```python
def batch_merge_folders(folder_specs):
    """
    Merge multiple folder sources.
    
    Args:
        folder_specs: List of dicts with 'input_folder' and 'output_file'
    
    Example:
        specs = [
            {'input_folder': '/path/to/folder1', 'output_file': 'merged_1.csv'},
            {'input_folder': '/path/to/folder2', 'output_file': 'merged_2.csv'},
        ]
        batch_merge_folders(specs)
    """
    for spec in folder_specs:
        print(f"\nProcessing: {spec['input_folder']}")
        print("=" * 50)
        merge_csv_files(
            spec['input_folder'],
            spec['output_file'],
            skip_pattern=spec.get('skip_pattern')
        )

# Example usage (uncomment to run)
# specs = [
#     {'input_folder': '/path/folder1', 'output_file': 'merged_1.csv', 'skip_pattern': 'merged'},
#     {'input_folder': '/path/folder2', 'output_file': 'merged_2.csv', 'skip_pattern': 'merged'},
# ]
# batch_merge_folders(specs)
```
